```python
import numpy as np
import scipy as sp

sample_rate = 32e3

with open('../data/signal.bin', 'r') as file:
    signal = sp.fromfile(file, dtype=sp.uint8)

time = np.arange(0, len(signal)) / sample_rate
```

```python
from matplotlib import pyplot as plt

plt.figure(dpi=300)
plt.axvline(1.31, color='orange')
plt.axvline(1.44, color='orange')
plt.plot(time, signal)
plt.show()
```

```python
from dsp import select_time_window

t, s = select_time_window(time, signal, 1.28, 1.46)

plt.figure(dpi=300)
plt.axvline(1.309, color='orange')
plt.axvline(1.313, color='orange')
plt.plot(t, s)
plt.show()
```
